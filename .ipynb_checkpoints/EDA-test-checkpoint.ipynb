After studying the dataset, I found interested to find out the relation between the price of the Airbnb properties and if they do or don't allow pets. 

For this, I want to check how many properties of the total allows pets.

```python
data['pets_allowed'].value_counts()
```

We can see that just around 5500 properties allows pets out of more than 30000

Below there is a plot to check the relation between the price and if the pets are allowed or not. 

```python
sns.boxplot(data=data, y='price', x="pets_allowed");
```

```python
sns.violinplot(data=data, y='price', x="pets_allowed");
```

1. We determinated that the properties accepting pets are slightly more expensive. 

This is probably because there are much less properties that accepts pets, so they charge more because of the extra cleaning service. 

I want to know the unique types of properties in the dataset to see if there is a relation between the type and the pets situation.

```python
property_types = data['property_type'].unique().tolist()
property_types
```

Now that we have all the types and we need to determinate the ammount of properties accepting pets per type.

For this, here is a for loop isolating both categories in order to be able to put them in a new dataframe.

```python
pet_allowed_counts = []
pet_not_allowed_counts = []

for _property in property_types:
    yes = len(data[(data['pets_allowed'] == 1) & (data['property_type'] == _property)])
    no = len(data[(data['pets_allowed'] == -1) & (data['property_type'] == _property)])
    pet_allowed_counts.append(yes)
    pet_not_allowed_counts.append(no)
```

```python
prop_pets = {'property_types': property_types,
             'pets_allowed': pet_allowed_counts,
             'pets_not_allowed': pet_not_allowed_counts
            }

property_pets_rel = pd.DataFrame(prop_pets)
property_pets_rel
```

This dataframe shows per property type, the ammount that are listing as pets allowed.

We can see that in all the types, the not allowing pets properties are much higher than the ones which do accept.

In order to have a better plot for this new dataframe, the row 'Apartments' has to be deleted.

```python
update_property_pets_rel = property_pets_rel.drop(0)
update_property_pets_rel
```

```python
import matplotlib.pyplot as plt

# plot bars in stack manner
plt.bar(update_property_pets_rel['property_types'], update_property_pets_rel['pets_allowed'], color='r')
plt.bar(update_property_pets_rel['property_types'], update_property_pets_rel['pets_not_allowed'], bottom=update_property_pets_rel['pets_allowed'], color='b')
plt.xticks(rotation = 45) 
plt.show()
```
